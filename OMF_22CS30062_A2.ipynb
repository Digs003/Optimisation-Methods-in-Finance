```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cov=pd.read_excel("covariance_matrix.xlsx")
```

```python
cov
```

```python
cov_decimal=cov.iloc[:,1:]/1e4 #Conversion from % to decimal values
cov_decimal
```

```python
dataset=pd.read_excel('/content/30_assets_5Years.xlsx')
```

```python
dataset
```

```python
return_column=dataset['Expected_Rate_return']
return_column=np.array(return_column)
return_column
mean_set=set(return_column)
mean_return=list(mean_set)
mean_return=np.array(mean_return)
mean_return
```

## Market Portfolio Calculation


```python
mu_f=0.01#Risk free return
e=np.ones((30,1))
A=mean_return-mu_f

cov_decimal_inv=np.linalg.inv(cov_decimal)
N=np.matmul(cov_decimal_inv,A)
D=np.matmul(e.T,N)
W=N/D

W ##Market Portfolio
```

```python
return_mt=np.matmul(mean_return,W)
print("Return of Market Portfolio: ",return_mt)
```

```python
risk=np.matmul(np.matmul(W.T,cov_decimal),W)
print("Risk of Market Portfolio: ",risk)
```

```python
sigma_mt=np.sqrt(risk)#(return_mt,sigma_mt) represents the market portfolio
```

## Capital Market Line & Market Point

```python
def CML(sigma):
  return mu_f+((return_mt-mu_f)/sigma_mt)*sigma
```

```python
sigma_vals=np.linspace(0.02,0.06,1000)
cml_returns=CML(sigma_vals)
```

```python
plt.figure(figsize=(10,6))
plt.plot(sigma_vals,cml_returns,label="Capital Market Line",color="red")
plt.scatter(sigma_mt,return_mt,label="Capital Market Point",color="blue",marker='o',s=100)
plt.xlabel("Risk")
plt.ylabel("Return")
plt.title("Capital Market Line & Capital Market Point")
plt.legend()
plt.grid(True)
plt.show()
```

## Security Market Line

```python
def individual_beta(mean):
  return (mean-mu_f)/(return_mt-mu_f)
```

```python
beta_assets=[]
for i in range(len(mean_return)):
  beta_assets.append(individual_beta(mean_return[i]))
```

```python
plt.figure(figsize=(10,6))
plt.plot(beta_assets,mean_return,label="Security Market Line",color="red")
plt.xlabel("Beta")
plt.ylabel("Return")
plt.title("Security Market Line")
plt.legend()
plt.grid(True)
plt.show()
```
